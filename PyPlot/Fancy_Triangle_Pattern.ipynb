```python
def is_point_inside_triangle(point, triangle):
    """
    Check whether a point is inside a triangle using barycentric coordinates.

    Parameters:
        point (tuple): A tuple of (x, y) coordinates for the point to check.
        triangle (tuple of tuples): A tuple of three tuples, each representing the (x, y) coordinates
            of a vertex of the triangle in clockwise order.

    Returns:
        bool: True if the point is inside the triangle, False otherwise.
    """
    ax, ay = triangle[0]
    bx, by = triangle[1]
    cx, cy = triangle[2]
    px, py = point

    # Calculate the barycentric coordinates of the point
    denominator = (by - cy)*(ax - cx) + (cx - bx)*(ay - cy)
    if denominator == 0:
        return False  # Triangle is degenerate (has zero area)
    u = ((by - cy)*(px - cx) + (cx - bx)*(py - cy)) / denominator
    v = ((cy - ay)*(px - cx) + (ax - cx)*(py - cy)) / denominator
    w = 1 - u - v

    # Check if the point is inside the triangle
    if u >= 0 and v >= 0 and w >= 0:
        return True
    else:
        return False
```

```python
import random
import matplotlib.pyplot as plt

# Main Points
A = [0, 0]
B = [1000, 1000]
C = [2000, 0]

# Point List
X = [A[0], B[0], C[0]]
Y = [A[1], B[1], C[1]]

# Random Point
point_0 = [-1, -1]
while not is_point_inside_triangle(point_0, (A, B, C)):
    point_0 = [random.randint(A[0]+1,C[0]), random.randint(A[0],B[1])]
X.append(point_0[0])
Y.append(point_0[1])

# Generate Points
for i in range(100000):
    Xt, Yt = random.choice((A, B, C))
    X.append(int((X[-1]+Xt)/2))
    Y.append(int((Y[-1]+Yt)/2))

# Create Plot
fig, axel = plt.subplots()
axel.scatter(X, Y, color='black', marker='^')
axel.set_title("The Sierpinski triangle")
```
